```python
# load a record using the 'rdrecord' function
record = wfdb.rdrecord("./data/two_lead/physionet_arrhythima/mit-bih-arrhythmia-database-1.0.0/100",
                       sampfrom=10000,
                       sampto=12000,
                        channels=[0,1])

# plot the record to screen
wfdb.plot_wfdb(record=record, title='Example signals')

print(type(record), record)
```
